```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Data generation
T = 10
M = 500
initial_values = [5, 80]
np.random.seed(222)

def generate_data(T, M, initial_values, a=0.3, b=80):
    for initial_value in initial_values:
        initial_value_1 = initial_value
        data = {0: initial_value}
        for t in range(1,M+1,1):
            initial_value += a * (b - initial_value) * T / M + np.random.normal(0, np.sqrt(T/M)) * np.sqrt(initial_value)
            data[t] = initial_value
        data = pd.DataFrame(list(data.items()), columns=['time', 'value'])
        data.set_index('time', inplace=True)
        filename = f'data{initial_value_1}.csv'
        print(data)
        data.to_csv(filename, index=False)
    return True

data_generated = generate_data(T, M, initial_values)
```

```python
data5 = pd.read_csv('data5.csv')
data80 = pd.read_csv('data80.csv')


sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
plt.plot(data5, label='Data Z(0)= 5', color='b', linewidth=2)
plt.xlabel('Index', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Generated Data Z(0)= 5', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(data80, label='Data Z(0)= 80', color='r', linewidth=2)
plt.xlabel('Index', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Generated Data Z(0)= 80', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
