```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import os
import re
from collections import defaultdict

N = 300
pmf_dir = '/Users/qianxinhui/Desktop/NU-Research/kellogg/change-point-detection/data_integrated/pmf_data/M=500'
pmf_files = [f for f in os.listdir(pmf_dir) if f.startswith('pmf_') and f.endswith('.csv')]

pattern = re.compile(r"^(.+)ws([\.0-9]+)\.csv$")
pmf_grouped_by_prefix = defaultdict(list)

for fname in pmf_files:
    m = pattern.match(fname)
    if m:
        prefix = m.group(1)
        ws_val = m.group(2)
        pmf_grouped_by_prefix[prefix].append((ws_val, fname))

hist_data_path = '/Users/qianxinhui/Desktop/NU-Research/kellogg/change-point-detection/data_integrated/Simulation_histograms'

theory_colors = ['red', 'darkgreen', 'orange', 'purple', 'brown']

for prefix, ws_fname_list in pmf_grouped_by_prefix.items():
    selected_window_sizes = ['2', '3', '4', '10']
    
    print(f"Processing prefix: {prefix}")
    
    filtered_files = [(ws_val, fname) for ws_val, fname in ws_fname_list 
                     if ws_val in selected_window_sizes]

    try:
        parts = prefix.split('_')
        t = int([p for p in parts if p.startswith('t')][0][1:])
        z_initial = int(parts[parts.index('value') + 1])
        service_rate = 10 if z_initial == 5 else 100 if z_initial == 80 else None
        if service_rate is None:
            print(f"Unknown z_initial: {z_initial}")
            continue
    except Exception as e:
        print(f"Failed to parse {prefix}: {e}")
        continue

    hist_filename = f"for_histogram_CoxM1_Z0{z_initial}_serv{service_rate}_t{t}.pickle"
    hist_path = os.path.join(hist_data_path, hist_filename)
    
    try:
        with open(hist_path, 'rb') as f:
            hist_data = pickle.load(f)
        counts = hist_data['counts']
        bins = hist_data['bins']
        if bins[0] != 0:
            bins = np.array(bins, dtype=int)
            interval = max(int(bins[1] - bins[0]), 1)
            bins_before = np.arange(0, bins[0], interval)
            bins = np.concatenate([bins_before, bins])
            counts_before = np.zeros(len(bins_before))
            counts = np.concatenate([counts_before, counts])
    except Exception as e:
        print(f" Failed to load histogram {hist_filename}: {e}")
        counts = bins = None

    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Plot normalized histogram
    if counts is not None and bins is not None:
        x_hist = bins[:-1]
        # Normalize counts to probabilities
        total_count = np.sum(counts)
        if total_count > 0:
            probabilities = counts / total_count
        else:
            probabilities = counts
            print(f"Warning: total_count is 0 for {prefix}")
        
        ax.bar(x_hist, probabilities, 
               width=bins[1] - bins[0], 
               alpha=0.4, 
               color='lightblue', 
               label=f'Simulation t={t}',
               zorder=1)
        
        print(f" Plotted histogram with {len(counts)} bins")
        print(f"  Total count: {total_count}")
        print(f"  Probability sum: {np.sum(probabilities):.6f}")
        print(f"  Histogram y-range: {probabilities.min():.6f} to {probabilities.max():.6f}")
    
    # Plot theory curves
    plotted_any_theory = False
    for idx, (ws_val, fname) in enumerate(sorted(filtered_files, key=lambda x: float(x[0]))):
        try:
            pmf_path = os.path.join(pmf_dir, fname)
            pmf = pd.read_csv(pmf_path).values.flatten()
            x_vals = np.arange(min(len(bins), N))
            y_vals = pmf[:min(len(bins), N)]
            
            color = theory_colors[idx] if idx < len(theory_colors) else f'C{idx}'
            ax.plot(x_vals, y_vals, 
                   marker='o', 
                   linestyle='-', 
                   color=color, 
                   label=f'Theory WS={ws_val}', 
                   linewidth=2, 
                   markersize=4,
                   zorder=2)
            
            print(f" Plotted theory curve: WS={ws_val}")
            print(f"  PMF sum: {sum(pmf):.6f}, Clipped PMF sum: {sum(y_vals):.6f}")
            print(f"  Theory y-range: {y_vals.min():.6f} to {y_vals.max():.6f}")
            
            plotted_any_theory = True
            
        except Exception as e:
            print(f" Failed to plot {fname}: {e}")
            continue
    
    if not plotted_any_theory:
        print(f" No theory curves plotted for {prefix}")
        plt.close(fig)
        continue
    
    # Set plot properties
    x_max = max(x_hist[-1], x_vals[-1]) + 1
    ax.set_xlim(0, x_max)
    ax.set_title(f'Queue Length Distribution: Cox/M/1\nt={t}, Z={z_initial}, ={service_rate}', 
                fontsize=14, fontweight='bold')
    ax.set_xlabel('Queue Length (Number of Customers)', fontsize=12)
    ax.set_ylabel('Probability', fontsize=12)
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)
    
    # Set x-axis limits based on meaningful data
    # if counts is not None and probabilities is not None:
    #     max_prob = max(probabilities) if len(probabilities) > 0 else 0
    #     meaningful_indices = np.where(probabilities > max_prob * 0.01)[0]
    #     if len(meaningful_indices) > 0:
    #         ax.set_xlim(0, min(meaningful_indices[-1] + 5, 40))
    
    # Set y-axis to start from 0
    ax.set_ylim(bottom=0)
    
    plt.tight_layout()
    plt.show()
    print(f" Completed: {prefix}")
    print("-" * 50)

print("Done!")
```
