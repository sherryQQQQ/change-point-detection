```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Data generation -> initial setup

T = 10
M = 500
initial_values = [5, 80]
np.random.seed(222)

def generate_data(T, M, initial_values, a=0.3, b=80):
    for initial_value in initial_values:
        initial_value_1 = initial_value
        data = {0: initial_value}
        for t in range(1,M+1,1):
            initial_value += a * (b - initial_value) * T / M + np.random.normal(0, np.sqrt(T/M)) * np.sqrt(initial_value)
            data[t * T/M] = initial_value  # Changed index to t * T/M
        data = pd.DataFrame(list(data.items()), columns=['time', 'value'])
        filename = f'initial_value_{initial_value_1}_samples_{M}.csv'
        print(data)
        data.to_csv(filename, index=False)  # Changed index=True to store time index
    return True

data_generated = generate_data(T, M, initial_values)
```

# visualization

```python
# M = 5000
data5 = pd.read_csv('initial_value_5_samples_5000.csv')
data80 = pd.read_csv('initial_value_80_samples_5000.csv')


sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
plt.plot(data5['time'], data5['value'], label='Data Z(0)= 5', color='b', linewidth=2)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Generated Data Z(0)= 5', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(data80['time'], data80['value'], label='Data Z(0)= 80', color='r', linewidth=2)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Generated Data Z(0)= 80', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
